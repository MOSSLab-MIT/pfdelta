```julia
# Check 2. see which cases have branches or generators off by result_default
net_57 = PM.parse_matpower("pglib/pglib_opf_case57_ieee.m")
net_118 = PM.parse_matpower("pglib/pglib_opf_case118_ieee.m")
net_500 = PM.parse_matpower("pglib/pglib_opf_case500_goc.m")
net_2000 = PM.parse_matpower("pglib/pglib_opf_case2000_goc.m")

nets = Dict(
    "case57" => net_57,
    "case118" => net_118,
    "case500" => net_500,
    "case2000" => net_2000,
)

for (name, net) in nets
    counter_off_branches = sum(branch["br_status"] == 0 for branch in values(net["branch"]))
    counter_off_gens = sum(gen["gen_status"] == 0 for gen in values(net["gen"]))

    println("Number of branches off for $name: $counter_off_branches")
    println("Number of generators off for $name: $counter_off_gens")
end
```

```julia
# Check 3. See if new topology perturbation function is working
net_500 = PM.parse_matpower("pglib/pglib_opf_case5_pjm.m")
net_500["branch"]["1"]["br_status"] = 0
net_500["gen"]["1"]["gen_status"] = 0
perturb_topology!(net_500; method="n-2")

# Print branch statuses sorted numerically by key
println("Branch statuses after perturbation:")
for (k, branch) in sort(collect(net_500["branch"]), by = x -> parse(Int, x[1]))
    println("  Branch $k: br_status = $(branch["br_status"])")
end

# Print generator statuses sorted numerically by key
println("\nGenerator statuses after perturbation:")
for (k, gen) in sort(collect(net_500["gen"]), by = x -> parse(Int, x[1]))
    println("  Gen $k: gen_status = $(gen["gen_status"])")
end
```
